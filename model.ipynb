```python
def realtime_detection(path_to_video: str, process: bool = True, model_type: str = 'YOLO'):
    capture = segmentator.get_capture(path_to_video)
    print(os.environ['VIDEO_FRAME_WIDTH'], os.environ['VIDEO_FRAME_HEIGHT'])

    while (capture.isOpened()):
        gc.collect()
        torch.cuda.empty_cache()

        ret, frame = capture.read()
        if ret == True:
            if process:
                frame = segmentator.apply_brightness_contrast_sharpening(frame, int(os.environ['CONVERT_BRIGHTNESS']) + 255,
                                                                                 int(os.environ['CONVERT_CONTRAST']) + 127,
                                                                                 int(os.environ['CONVERT_SHARPENING_CYCLE']), prod=True,
                                                                                 process=process, resize=False)

            cv2.imwrite(os.environ['TEMP_IMAGE_FILE'], frame)

            segmentator.predict_dataset(os.environ['TEMP_IMAGE_FILE'], model_type)
            print('temp im fil', os.environ['TEMP_LABEL_FILE'])
            if os.path.exists(os.environ['TEMP_LABEL_FILE']):
                print('exist')
                markup = segmentator.handle_file(os.environ['TEMP_LABEL_FILE'], stream=True)
                markup['image_path'] = os.environ['PREDICTED_IMAGE_PATH']

                os.remove(os.environ['TEMP_LABEL_FILE'])

                segmentator.add_markup(markup, True)


            image = cv2.imread(os.environ['PREDICTED_IMAGE_PATH'], cv2.IMREAD_COLOR)
            cv2.imshow('Frame', image)

            # add waitKey for video to display
            cv2.waitKey(1)

            if cv2.waitKey(25) == ord('q'):
                # do not close window, you want to show the frame
                # cv2.destroyAllWindows();
                cv2.destroyAllWindows()
                segmentator.clear_cache()
                break
        else:
            break

from model import ODaSModel

segmentator = ODaSModel()
segmentator.setup_env()
segmentator.load_model("YOLO", use_default_model=True)

process = False
path_to_video = "./Download.mp4"

realtime_detection(path_to_video, process)
```

```python
def calculate_polygon_area(x,y):
    return 0.5*np.abs(np.dot(x,np.roll(y,1))-np.dot(y,np.roll(x,1)))

def handle_file(filepath):
    f = open(filepath, "r")
    lines = f.readlines()

    num_of_units = len(lines)

    line_with_max_square_index = -1
    max_square = -1
    best_line_coords = []

    # посчитали
    for i, line in enumerate(lines):
        coords = list(map(float, line.split()[1:]))

        x_coords = np.array(coords[::2]) * int(os.environ['IMAGE_SIZE'])
        y_coords = np.array(coords[1::2]) * int(os.environ['IMAGE_SIZE'])

        sqaure = calculate_polygon_area(x_coords, y_coords)

        if sqaure > max_square:
            max_square = sqaure
            line_with_max_square_index = i
            best_line_coords = dict({
                'x' : x_coords,
                'y' : y_coords
            })

    res = dict({
        'num_of_units' : num_of_units,
        'line_with_max_square_index' : line_with_max_square_index,
        'max_square' : max_square,
        'best_line_coords' : best_line_coords
    })

    return res

# def get_all_coords(filepath):
#     res = []
#
#     f = open(filepath, "r")
#     lines = f.readlines()
#
#     for i, line in enumerate(lines):
#         coords = list(map(float, line.split()[1:]))
#         x_cords = coords[::2]
#         y_cords = coords[1::2]
#
#         polygon = dict({
# #             'index': i,
#             'x' : x_cords,
#             'y' : y_cords
#         })
#
#         res.append(polygon)
#
#     return res

def handle_labels_dir(dirpath):
    res = []
    onlyfiles = [f for f in listdir(dirpath) if isfile(join(dirpath, f))]

    for filename in onlyfiles:
        fileres = handle_file(os.path.join(dirpath, filename))
        fileres['file_name'] = filename
        image_path = os.path.join(os.environ['PREDICTED_DATA_PATH'], filename.replace(".txt", ".jpg"))
        if os.path.exists(image_path):
            fileres['image_path'] = image_path
        else:
            raise FileNotFoundError('label exists, image is not')
        res.append(fileres)

    return res


def add_markups(markups):
    
    DPI = 25
    w = int(os.environ['IMAGE_SIZE']) / (DPI*10)
    h = int(os.environ['IMAGE_SIZE']) / (DPI*10)
    
    for markup in markups:
        data = img.imread(markup['image_path'])

        x = markup['best_line_coords']['x'].astype(int)
        y = markup['best_line_coords']['y'].astype(int)

        fig = plt.figure(frameon=False, figsize=(w, h), dpi=DPI)

        ax = plt.Axes(fig, [0., 0., 1., 1.])
        ax.set_axis_off()
        fig.add_axes(ax)

        ax.text(270, 50, f"Est. biggest size: {round(markup['max_square'], 1)} pixels",
                 color="white",
             horizontalalignment='center',
             verticalalignment='center')

        ax.plot(x, y, color="white", linewidth=2)

        ax.imshow(data, aspect='auto')
        fig.savefig(markup['image_path'], dpi=DPI*10)

def proceed_info():
    markups = handle_labels_dir(os.path.join(os.environ['PREDICTED_DATA_PATH'], 'labels'))
    add_markups(markups)
```
