```python
def polygon_area(x,y):
    # coordinate shift
    x_ = x - x.mean()
    y_ = y - y.mean()
    # everything else is the same as maxb's code
    correction = x_[-1] * y_[0] - y_[-1]* x_[0]
    main_area = np.dot(x_[:-1], y_[1:]) - np.dot(y_[:-1], x_[1:])
    return 0.5*np.abs(main_area + correction)
```

```python
def PolyArea(x,y):
    return 0.5*np.abs(np.dot(x,np.roll(y,1))-np.dot(y,np.roll(x,1)))

def handle_file(filepath):
    f = open(filepath, "r")
    lines = f.readlines()
    
    num_of_units = len(lines)
    
    line_with_max_square_index = -1
    max_square = -1
    best_line_coords = []
    
    # посчитали
    for i, line in enumerate(lines):
        coords = list(map(float, line.split()[1:]))
        x_cords = coords[::2]
        y_cords = coords[1::2]
    
        sqaure = PolyArea(x_cords, y_cords)
        
        if sqaure > max_square:
            max_square = sqaure
            line_with_max_square_index = i
            best_line_coords = dict({
                'x' : x_cords,
                'y' : y_cords
            })
            
    best_line = lines[line_with_max_square_index]     
    
    res = dict({
        'num_of_units' : num_of_units,
        'line_with_max_square_index' : line_with_max_square_index,
        'max_square' : max_square,
        'best_line_coords' : best_line_coords
    })
    
    return res

def get_all_coords(filepath):
    res = []
    
    f = open(filepath, "r")
    lines = f.readlines()
    
    for i, line in enumerate(lines):
        coords = list(map(float, line.split()[1:]))
        x_cords = coords[::2]
        y_cords = coords[1::2]
        
        polygon = dict({
#             'index': i,
            'x' : x_cords,
            'y' : y_cords
        })
        
        res.append(polygon)
        
    return res

def handle_dir(dirpath):
    res = []
    onlyfiles = [f for f in listdir(dirpath) if isfile(join(dirpath, f))]
    
    for filename in onlyfiles:
        fileres = handle_file(os.path.join(dirpath, filename))
        fileres['filename'] = filename
        res.append(fileres)
        
    return res
```
